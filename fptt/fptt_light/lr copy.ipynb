```python
def errorfill(x, y, yerr, color=None, alpha_fill=0.3, ax=None,label=None):
    y = np.array(y)
    yerr = np.array(yerr)
    ax = ax if ax is not None else plt.gca()
    if color is None:
        color = ax._get_lines.color_cycle.next()
    if np.isscalar(yerr) or len(yerr) == len(y):
        ymin = y - yerr
        ymax = y + yerr
    elif len(yerr) == 2:
        ymin, ymax = yerr
    ax.plot(x, y, color=color,label=label)
    ax.fill_between(x, ymax, ymin, color=color, alpha=alpha_fill)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
ind = np.array([20,40,60,80,100,200,500])   # the x locations for the groups

errorfill(ind, bptt_snn_aux_mean, yerr= bptt_snn_aux_std,color = 'g', label='BPTT-SNN-Aux')
errorfill(ind, bptt_lstm_aux_mean,yerr= bptt_lstm_aux_std,color = 'b', label='BPTT-LSTM-Aux')
errorfill(ind, fptt_snn_mean,  yerr= fptt_snn_std, color = 'r',label='FPTT-SNN')


ax.set_ylabel('Accuracy %')
ax.set_xlabel('Frames')
ax.set_ylim(0,95.)
ax.set_xticks(ind+width*2)
ax.set_xticklabels(('20', '40', '60', '80','100','200','500'))
ax.legend((p1[0],p2[0],p3[0],p4[0],p5[0]),('BPTT-SNN-Aux','BPTT-LSTM-Aux','FPTT-SNN'),ncol=5,
            bbox_to_anchor=(1.005, 1.08), prop={'size': 12.5})
plt.show()
```
